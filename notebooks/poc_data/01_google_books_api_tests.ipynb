## Import

```python
isbns = [
    "9782351423554",
    "9791042019396",
    "9782820354488",
    "9782290042298",
    "9782845808331",
    "9782413042341",
    "9782384967421",
    "9791043301087",
    "9782374123035",
    "9782374126173",
    "9791039143431",
    "9791026854920",
    "9791039140652",
    "9791026856283",
    "9791039147156",
    "9782203001190",
    "9782858150045",
    "9791038206229",
    "9782822244787",
    "9791038209657",
]
```

## Fonction de test google books

```python

def build_google_books_results(isbns: list[str]) -> pd.DataFrame:
    """
    Lance un test Google Books sur une liste d'ISBN/EAN.

    La fonction capture les erreurs afin de ne pas interrompre le notebook.

    Args:
        isbns: Liste d'ISBN/EAN à tester.

    Returns:
        DataFrame contenant les résultats normalisés et les erreurs éventuelles.
    """
    rows = []

    for isbn in isbns:
        try:
            result = search_book_by_isbn(isbn)

            if result is None:
                rows.append(
                    {
                        "source": "google_books",
                        "isbn_query": isbn,
                        "found": False,
                        "error": "Aucun résultat ou quota/API indisponible",
                    }
                )
            else:
                rows.append(
                    {
                        "source": "google_books",
                        "isbn_query": isbn,
                        "found": True,
                        "error": None,
                        **result,
                    }
                )

        except Exception as error:
            rows.append(
                {
                    "source": "google_books",
                    "isbn_query": isbn,
                    "found": False,
                    "error": str(error),
                }
            )

    return pd.DataFrame(rows)
```

## Exécution du test

```python

df_google = build_google_books_results(isbns)

df_google
```

## Complétude des métadonnées

```python

metadata_fields = [
    "title",
    "authors",
    "publisher",
    "published_date",
    "language",
    "description",
    "categories",
    "thumbnail",
]

available_fields = [
    field for field in metadata_fields
    if field in df_google.columns
]

metadata_completeness = (
    df_google[available_fields]
    .notna()
    .mean()
    .mul(100)
    .round(2)
    .reset_index()
)

metadata_completeness.columns = ["field", "completion_rate"]

metadata_completeness
```

```python
df_google_ok = df_google[df_google["found"]].copy()
df_google_ko = df_google[~df_google["found"]].copy()

display(df_google_ok)
display(df_google_ko)
```


## Premiers constats

Les premiers tests réalisés sur Google Books API montrent plusieurs limites importantes dans le cadre du POC CollectionLens.

### Couverture des ouvrages

Le taux de couverture observé sur l’échantillon testé est d’environ 50 %, ce qui signifie qu’une partie importante des ISBN manga, BD et comics n’est pas retrouvée.

Les ouvrages récents ou issus d’éditions françaises semblent particulièrement moins bien couverts.

### Complétude des métadonnées

Les champs suivants sont relativement bien renseignés :
- titre ;
- auteurs ;
- date de publication ;
- catégories.

En revanche, plusieurs métadonnées restent très incomplètes :
- éditeur ;
- description ;
- miniature (thumbnail).

### Limites identifiées

Les limites suivantes ont été observées :
- couverture incomplète des ISBN ;
- qualité variable des métadonnées ;
- absence fréquente des descriptions ;
- quotas journaliers Google Books API ;
- faible couverture de certains mangas, BD et comics français.

### Conclusion POC

Google Books API constitue une source intéressante pour enrichir certaines métadonnées bibliographiques, mais ne semble pas suffisante comme source unique pour CollectionLens.

Ces résultats justifient :
- l’évaluation d’autres APIs bibliographiques ;
- la mise en place d’une stratégie multi-sources ;
- l’ajout futur de mécanismes de cache et de normalisation.
