```python
from pathlib import Path
import sys
import json
from pprint import pprint

PROJECT_ROOT = Path.cwd().parents[1]
sys.path.append(str(PROJECT_ROOT / "src"))

from collectionlens.pipelines.metadata_aggregation import (
    aggregate_book_metadata,
)
```

```python
RAW_DIR = PROJECT_ROOT / "data" / "raw" / "pipeline_quality_baseline"

isbn = "9782351423554"

sources_data = {}

for source_name in [
    "nudger",
    "google_books",
    "bnf",
    "openlibrary",
]:
    json_path = RAW_DIR / source_name / f"{isbn}.json"

    with json_path.open(
        "r",
        encoding="utf-8",
    ) as file:
        payload = json.load(file)

    sources_data[source_name] = payload["result"]
```

```python
aggregated_book = aggregate_book_metadata(
    sources_data=sources_data,
)

aggregated_book
```
